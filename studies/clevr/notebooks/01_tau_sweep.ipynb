# 01 — Softmin Temperature (τ) Sweep

## Hypothesis

The softmin temperature τ controls the sharpness of the attention-like weights used by Power-SM3 to assign each predicted slot to its nearest ground-truth object.

- **Too low** (τ → 0): weights collapse to a hard nearest-neighbour assignment. Gradients are blocked from all non-nearest slots, making optimisation brittle and potentially harming coverage.
- **Too high** (τ → ∞): weights become uniform across all slots. Gradient signals from many object pairs cancel, slowing convergence and blurring the loss landscape.
- **Optimal τ**: there exists a sweet spot where the loss landscape is smooth enough to optimise yet sharp enough to maintain meaningful slot-to-object alignment.

**Experiment**: train Power-SM3 with τ ∈ {0.07, 0.10, 0.12, 0.15, 0.20, 0.30} on the CLEVR dataset and compare validation metrics and PCA slot-embedding snapshots across training.

```python
import sys
sys.path.insert(0, '.')
import clevr_utils as cu
from influencerformer.losses import PowerSoftMinLoss
print(f'Device: {cu.DEVICE}')
```

```python
TAU_LIST = [0.07, 0.10, 0.12, 0.15, 0.20, 0.30]
N_EPOCHS = 50
SNAPSHOT_EPOCHS = {0, 5, 15, 35, 49}
```

```python
all_results = {}

for tau in TAU_LIST:
    loss_fn = PowerSoftMinLoss(temperature=tau, power=3.0)
    label = f'τ={tau}'
    print(f'\n=== Training {label} ===')
    metrics, snapshots = cu.train_and_monitor(
        loss_fn,
        loss_name=label,
        n_epochs=N_EPOCHS,
        snapshot_epochs=SNAPSHOT_EPOCHS,
        tau_for_entropy=tau,
    )
    all_results[label] = (metrics, snapshots)
```

```python
cu.plot_monitoring(all_results)

best_tau_label = min(
    all_results,
    key=lambda k: min(all_results[k][0].get('val_loss', [float('inf')]))
)
print(f'Best τ by val_loss: {best_tau_label}')
best_metrics, best_snapshots = all_results[best_tau_label]
cu.plot_pca_snapshots(best_snapshots, title=f'PCA Slot Embeddings — {best_tau_label}')
```
